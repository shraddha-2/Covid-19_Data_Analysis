```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%d-%m-%Y')
```

```python
covid_df.head()
```

Active cases of covid 19 india 
Active cases = Total cases - Cured cases

```python
covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
```

to add new column active cases

```python
#Display last five rows
covid_df.tail() 
```

sum of confirmed, deaths and cured cases with index as State/UnionTerritory for each country
use pivot_table() function

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"],
                          index = "State/UnionTerritory", aggfunc = max)
```

```python
statewise
```

to find recovery rate of each state
Recovery rate = (cured cases * 100) / confirmed cases

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Recovery Rate"]
```

state_recovery_rate = statewise["Recovery Rate"]
state_recovery_rate.to_frame().style.background_gradient(cmap = "cubehelix")

to find mortaility rate of each state 
Mortality rate = (death cases*100)/confirmed cases

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
#Top ten states with active cases
top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by = ['Active_cases'], ascending = False).reset_index()
top_10_a_c = top_10_active_cases.iloc[:10]
top_10_a_c
fig = plt.figure(figsize = (16,9))
sns.set_theme(style = 'whitegrid')
ax = sns.barplot(data = top_10_a_c, x = 'State/UnionTerritory', y = 'Active_cases',
                linewidth = 2, edgecolor = 'black')

plt.title('Top 10 states with most active cases in India', size = 22)
plt.xlabel("States")
plt.ylabel("Total active cases")
```

```python
#Top ten states with highest deaths
top_10_deaths_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
top_10_d_c = top_10_deaths_cases.iloc[:12]

fig = plt.figure(figsize = (16,9))
sns.set_theme(style = 'whitegrid')
ax = sns.barplot(data = top_10_d_c, x = 'State/UnionTerritory',
                y = 'Deaths', linewidth = 2, edgecolor = 'black')

plt.title('Top 10 states with most deaths cases in India', size = 22)
plt.xlabel('Sates')
plt.ylabel('Total death cases')
```
